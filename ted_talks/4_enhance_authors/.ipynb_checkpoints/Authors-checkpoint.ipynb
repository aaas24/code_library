## Using BeautifulSoup Library

```python
page='List of TED speakers'

url=(page).replace(" ", "_")
url= 'https://en.wikipedia.org/wiki/'+ url

url
```

```python
#using beautiful soup to dig into the html
from bs4 import BeautifulSoup
from urllib.request import urlopen


html = urlopen (url)
bsObj = BeautifulSoup(html.read (), 'html.parser')
```

```python
bsObj
```

```python
def tag_list(tag):
    """
    This function extracts the list of tags and returns the list
    """
    soup=bsObj
    a=[]
    content=soup.findAll(tag)
    for item in content:
        a.append(item.get_text())
    return a
```

```python
authors= tag_list("td")
authors=(
    authors[6:]
)
authors
```

```python
#create dataframe from previous result & infer potential URL based on name

def custom_convert(x):
    y=str(x).lower().replace("/n"," ")
    return y

dic_authors_talks={}
start=0

i=0
for line in authors: 
    if '\n' in line:
        key=authors[i-1]
        value=authors[i].replace("\n", "")
        dic_authors_talks.update({key:value})
    i=i+1

df=(pd.DataFrame.from_dict(dic_authors_talks, orient='index')
    .reset_index()
   )
df.columns=['speaker', 'talks']
df['potential_url']=('https://en.wikipedia.org/wiki/'+ df['speaker']).str.replace(" ", "_")
pd.set_option('display.max_colwidth', None)
df['name_code']=df['speaker'].apply(custom_convert)
df
```

```python
name="Tom"
q=["name_code.str.contains(","'",name.lower(),"'",")"]
q="".join(q)
result=(df
           .query(q)
)
result
```

```python
#obtain speaker url from HTML
import re

def custom_convert(x):
    y=str(x).lower().replace("/n"," ")
    return y
    
#Parse HTML to find urls of speakers 
result=[]
txt = str(bsObj)
# reg='(?<=a href=)(.*)(?=title)'
reg='(?<=data-sort-value=)(.*)(?=title)'

reobj = re.compile(reg)
for matchobj in reobj.finditer(txt):
    result.append(matchobj[1 ])
    
df_result=(pd.DataFrame(result)
           .iloc[5:937,:]
           .set_axis(['value'], axis=1)
           .value.str.split('!"><a href=', expand=True)
           .set_axis(['name', 'url'], axis=1)
           .reset_index(drop=True)
          )
df_result.url=df_result.url.str[1:-2].reset_index(drop=True)

df_result1=(df_result.name
                .str[1:]
                .str.split(',', expand=True)
                .set_axis(['last', 'first'], axis=1)
               )
df_result=df_result.drop(['name'], axis=1)
df_result['name']=df_result1['first']+ df_result1['last']
df_result['name_code']=df_result['name'].apply(custom_convert)


df_names=df_result.copy()
print(df_names) 
```

```python
name="Tom"
q=["name_code.str.contains(","'",name.lower(),"'",")"]
q="".join(q)
result=(df_names
           .query(q)
)
result
```
