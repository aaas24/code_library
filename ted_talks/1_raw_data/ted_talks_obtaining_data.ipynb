### Joining both dataframes obtained

```python
import os
cwd=os.getcwd()
df_scrapped=pd.read_csv(cwd+"/preprocessing/scrapped.csv")
df_kaggle=pd.read_csv(cwd+"/preprocessing/kaggle.csv")
```

```python

print(
"kaggle: ",df_kaggle.shape,'/n',
"scrapped: ",df_scrapped.shape
)
```

```python
df_scrapped=df_scrapped.drop(['Unnamed: 0'], axis=1)
df_scrapped.head(5)
```

```python
df_kaggle=df_kaggle.drop(['Unnamed: 0'], axis=1)
df_kaggle.head(5)
```

```python
df=df_kaggle.merge(df_scrapped, left_on='link', right_on='link')
df.to_csv(cwd+"/preprocessing/final_raw_data.csv")
```
