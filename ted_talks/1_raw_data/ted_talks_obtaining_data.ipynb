### Joining both dataframes obtained

```python
import os
cwd=os.getcwd()
cwd
```

```python
df_scrapped=pd.read_csv(cwd+"/../2_preprocessing/scrapped.csv")
df_kaggle=pd.read_csv(cwd+"/../2_preprocessing/kaggle.csv")
```

```python

print(
"kaggle: ",df_kaggle.shape,'/n',
"scrapped: ",df_scrapped.shape
)
```

```python
df_scrapped=df_scrapped.drop(['Unnamed: 0'], axis=1)
df_scrapped.head(5)
```

```python
df_kaggle=df_kaggle.drop(['Unnamed: 0'], axis=1)
df_kaggle.head(5)
```

```python
df=df_scrapped.merge(df_kaggle, left_on='link', right_on='link')
df.to_csv(cwd+"/final_raw_data.csv")
```
